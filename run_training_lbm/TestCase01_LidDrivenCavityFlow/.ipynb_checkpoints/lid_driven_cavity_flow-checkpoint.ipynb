# Cas test 'Lid driven cavity flow' en LBM BGK pour $Re = 100, 400, 1000$

## Cadre

Le 'Lid driven cavity flow' est un cas test classique en CFD.

Le domaine est une cavité carrée en deux dimensions. 
Ce cas test consiste à imposer une vitesse tangentielle $U$ le long de la paroi supérieure. 
Ce mouvement induit un vortex au centre de la cavité, tout le fluide se met à tourner autour.

Les profils de vitesse pour différents nombre de Reynolds sont comparés aux résultats d'un article de Ghia et al. [1].
